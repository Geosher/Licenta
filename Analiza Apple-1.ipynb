```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st

df = pd.read_csv("Apple.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df.groupby('Year')['Close'].mean().plot(kind='line', title="Prețul mediu anual de închidere")
```

```python
df.groupby('Year')['Volume'].mean().plot(kind='bar', figsize=(10,5), title="Volumul mediu anual tranzacționat")
```

```python
df['Month'] = df['Date'].dt.month
df.groupby('Month')['Close'].mean().plot(kind='line', title="Preț mediu pe lună")
```

```python
df['Volatility'] = df['High'] - df['Low']
df.groupby('Year')['Volatility'].mean().plot(title="Volatilitate anuală medie")
```

```python
df['Pct_Change'] = df['Close'].pct_change()
df['Pct_Change'].plot(title="Variația zilnică procentuală a prețului")
```

```python
st.title("Analiza acțiunilor Apple")

df = pd.read_csv("Apple.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

annual_price = df.groupby('Year')['Close'].mean()

fig, ax = plt.subplots()
ax.plot(annual_price.index, annual_price.values, marker='o')
ax.set_title("Prețul mediu anual de închidere")
ax.set_xlabel("Anul")
ax.set_ylabel("Preț (USD)")

st.pyplot(fig)
```
