```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


apple_df = pd.read_csv("apple_prelucrat.csv")

```

```python
plt.figure(figsize=(12,6))
plt.plot(apple_df['Date'], apple_df['Close'], label='Preț închidere', color='blue')
plt.title("Evoluția prețului acțiunii Apple în timp")
plt.xlabel("Dată")
plt.ylabel("Preț de închidere ($)")
plt.legend()
plt.tight_layout()
plt.savefig("evolutie_close_timp.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(apple_df['Close'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribuția prețului de închidere")
plt.xlabel("Preț ($)")
plt.ylabel("Frecvență")
plt.tight_layout()
plt.savefig("distributie_close.png")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=apple_df, x='Month', y='Close', palette='Blues')
plt.title("Distribuția prețului de închidere pe luni")
plt.xlabel("Lună")
plt.ylabel("Preț ($)")
plt.tight_layout()
plt.savefig("boxplot_lunar_close.png")
plt.show()
```

```python
volatility_by_month = apple_df.groupby('Month')['Volatility'].mean()

plt.figure(figsize=(10,5))
volatility_by_month.plot(kind='bar', color='orange')
plt.title("Volatilitate medie lunară")
plt.xlabel("Lună")
plt.ylabel("Volatilitate ($)")
plt.tight_layout()
plt.savefig("volatilitate_medie_lunara.png")
plt.show()
```

```python
apple_df.groupby('Year')['Close'].mean().plot(kind='bar', figsize=(10,5), title='Preț mediu anual acțiune Apple')
```

```python

apple_df[['Close', 'Volume', 'Volatility']].describe()
```

```python
apple_df[apple_df['Year'] >= 2000].groupby('Year')['Close'].mean().plot(kind='bar', figsize=(10,5), title='Preț mediu anual Apple (din 2000)', color='lightgreen')
plt.ylabel('Preț ($)')
plt.tight_layout()
plt.savefig('pret_mediu_anual_din_2000.png')
plt.show()
```

```python
apple_df[apple_df['Year'] >= 2000].groupby('Year')['Volatility'].mean().plot(kind='line', figsize=(10,5), marker='o', title='Volatilitate medie anuală Apple (din 2000)')
plt.ylabel('Volatilitate ($)')
plt.tight_layout()
plt.savefig('volatilitate_medie_din_2000.png')
plt.show()
```
