```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

apple_df = pd.read_csv('apple_prelucrat.csv')
```

```python

print(apple_df[['Close', 'Volume', 'Volatility']].describe())
```

```python
plt.figure(figsize=(10,5))
plt.hist(apple_df['Close'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribuția prețului de închidere")
plt.xlabel("Preț ($)")
plt.ylabel("Frecvență")
plt.tight_layout()
plt.savefig("distributie_close.png")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(apple_df['Date'], apple_df['Close'], label='Preț închidere', color='blue')
plt.title("Evoluția prețului acțiunii Apple în timp")
plt.xlabel("Dată")
plt.ylabel("Preț de închidere ($)")
plt.legend()
plt.tight_layout()
plt.savefig("evolutie_close_timp.png")
plt.show()
```

```python
print(apple_df[['Date', 'Close']].head())
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=apple_df, x='Month', y='Close', palette='Blues')
plt.title("Distribuția prețului de închidere pe luni")
plt.xlabel("Lună")
plt.ylabel("Preț ($)")
plt.tight_layout()
plt.savefig("boxplot_lunar_close.png")
plt.show()
```
